```python
import sys
import matplotlib.pyplot as plt
import numpy as np
import scipy
from scipy.integrate import solve_ivp

print (sys.version)
```

```python
### setup ###

M_pl = 2.176e-8 #kg

V_0 = 4e-12
n = 15
lambda_v = 1e-9

#set 1
A = 1.335e-5
phi0_mpl = 2.527
sigma = 1e-3

def pd(xl, yl, t):
    sc = 8
    font_size = 20
    plt.figure(figsize=(sc * 2, sc * 1))
    plt.xlabel(xl, fontsize = font_size)
    plt.ylabel(yl, fontsize = font_size)
    plt.title(t, fontsize = font_size)
```

```python
#phi
phi_mpl = np.linspace(0, 15, 1000) #phi / M_pl, dimensionless

V_typ = V_0 * np.exp(-lambda_v * np.power(phi_mpl, n))

f_bump = A * np.exp(-0.5 * (np.power(phi_mpl - phi0_mpl, 2)) / (sigma**2))

# V(phi)
V_phi = V_typ * (1 + f_bump)

#plotting details
pd(r"$\phi$", r"$V(\phi)$", "")
plt.xlim([0, 15])
plt.ylim([0, 4.5e-12])

#plot!
plt.plot(phi_mpl, V_phi, color = "blue")
print(V_phi[1])
```
