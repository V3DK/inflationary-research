```python
import sys
import matplotlib.pyplot as plt
import numpy as np
import scipy
from scipy.integrate import solve_ivp

print (sys.version)
```

```python
### setup ###

#table 1 parameters (all a's are a tildes)
xi = 0.28
a2 = 2.18
a3 = -2.49
a5 = -6.22e-2
a6 = 2.20e-3
a4M2Mpl2 = 2.66e-10

def pd(xl, yl, t):
    sc = 8
    font_size = 20
    plt.figure(figsize=(sc * 2, sc * 1))
    plt.xlabel(xl, fontsize = font_size)
    plt.ylabel(yl, fontsize = font_size)
    plt.title(t, fontsize = font_size)
```

```python
phi = np.linspace(0, 10, 1000) #units: reduced plank mass, M_pl
V_t = (a4M2Mpl2 / np.power(1+xi*np.power(phi, 2), 2)) * (a2*np.power(phi, 2) + a3*np.power(phi, 3) + np.power(phi, 4) + a5*np.power(phi, 5) + a6*np.power(phi, 6))

#plotting details
pd(r"$\phi$", r"$\tilde{V}(\phi)$", "")
plt.xlim([0, 10])
plt.ylim([0, 1.2e-9])

#plot!
plt.plot(phi, V_t, color = "black")
```

```python
V_tp = np.gradient(V_t, phi[1]-phi[0])
V_tp2 = np.gradient(V_tp, phi[1]-phi[0])

#plotting details
pd(r"$\phi$, $[\bar{M}_{Pl}]$", r"$\tilde{V}(\phi)$", "")
plt.xlim([0, 10])
plt.ylim([-40e-11, 30e-11])

#plot!
plt.plot(phi, V_tp, "r")
plt.plot(phi, V_tp2, "b--")
```
