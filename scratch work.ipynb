```python
#test example, simple second order

w = 1
x_init = 0
vel = 1

def derivs(t, y):
    x = y[0]
    v = y[1]
    return [v, -(w**2 - (1 / t**2)) * x]

ts = np.logspace(0, 2, 1000)

sols = scipy.integrate.solve_ivp(derivs, t_span = [ts[0], ts[-1]], y0 = [x_init, vel], t_eval = ts)
# print(sols.y)
plt.plot(sols.t, sols.y[0])
plt.xscale('log', base = 10)
# plt.plot(sols.t, sols.y[1], color = "red")
```
